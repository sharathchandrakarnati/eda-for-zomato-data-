```python
df_country=pd.read_excel('Country-code.xlsx')
df_country
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how = 'left')
```

```python
final_df
```

```python
## to check data type
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values= final_df.Country.value_counts().values
```

```python
## pie chart-top3 country use zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

observation: zomato maximum records r transcations are from india after thst usa and then unitedkingdom

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating_count"})
```

```python
ratings
```

## observations
1. when rating is between 4.5 to 4.9---> Excellent 
2. when rating is between 4.0 to 4.4 ---> very good
3. when rating is between 3.5 to 3.9 --->good
4. when rating is between 2.5 to 3.4 ---> average 
5. when rating is between 1.8 to 2.4 ---> poor
6. when rating is between 0.0 ---> no ratings

```python

sns.barplot(x="Aggregate rating",y="Rating_count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating_count",hue='Rating color',data=ratings)
```

```python
sns.countplot(x="Rating color",data=ratings)
```
